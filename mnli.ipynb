```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification

tokenizer = AutoTokenizer.from_pretrained("facebook/bart-large-mnli")

model = AutoModelForSequenceClassification.from_pretrained("facebook/bart-large-mnli")
```

```python
#mkdir sam
save_path = '/content/0sam'
```

```python
def save_model(save_path):
        model.save_pretrained(save_path)
        tokenizer.save_pretrained(save_path)
```

```python
save_model(save_path)
```

```python
def preprocess_function(examples):
    hypothesis = examples['Hypothesis']
    premise = examples['Premise']
    label = examples['label']
    inputs = tokenizer(
        
        hypothesis,
        premise,
        max_length=384,
        truncation="only_second",
        return_offsets_mapping=True,
        padding="max_length",
    )
    inputs['label'] = label
    return inputs
```

```python
tokenized_anli = dataset.map(preprocess_function, batched=True)
```

```python
tokenized_anli1 = dataset1.map(preprocess_function, batched=True)
```

```python
from transformers import default_data_collator
data_collator = default_data_collator
```

```python
from transformers import TrainingArguments, Trainer
```

```python
import os 
```

```python
!pip install pytorch
```

```python
training_args = TrainingArguments(
    output_dir='./distil-roberta-base-zsl',
    #trainer = Trainer(default_root_dir="/your/path/to/save/checkpoints"),
    evaluation_strategy="epoch",
    learning_rate=5e-5,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=3,
    weight_decay=0.01,
)
```

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_anli["train"],
    
    eval_dataset=tokenized_anli1["train"],
    data_collator=data_collator,
    tokenizer=tokenizer,

)
```

```python
trainer.train()
```

```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification", model='/content/0sam')

sent = "Apple just announced the newest iPhone X"
candidate_labels = ["technology", "sports", "politics"]
res = classifier(sent, candidate_labels)
print(res)
```

```python
predictions = trainer.predict(tokenized_anli["train"])
```

```python
import numpy as np

preds = np.argmax(predictions.predictions, axis=-1)
from sklearn.metrics import classification_report
print(classification_report(predictions.label_ids, preds, digits=3))
```

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score
from transformers import EarlyStoppingCallback

accuracy = accuracy_score(y_true=predictions.label_ids, y_pred=preds)
recall = recall_score(y_true=predictions.label_ids, y_pred=preds,average='micro')
precision = precision_score(y_true=predictions.label_ids, y_pred=preds,average='micro')
f1 = f1_score(y_true=predictions.label_ids, y_pred=preds,average='micro')

print(accuracy) 
print(precision) 
print(recall)
```
