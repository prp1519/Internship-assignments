```python
from nltk.tokenize import word_tokenize 
from nltk.probability import FreqDist 
N = 250
tok1=[]
for i in range(len(df.Content)):
  text  = "{s}".format(s=df.Content[i])
  abcde = text.split(" ")
  new = word_tokenize(text)
  fdist = FreqDist(abcde)
  #fdist
  abcd = fdist.N()
  #abcd
  if (abcd) > N:
    new_text = new[:250]
    #print(250)
  else:
    new_text = new[:abcd]
    #print(abcd)
  lst = ' '.join([str(elem) for elem in new_text])
  tok1.append(lst)
```
