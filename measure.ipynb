```python
lis = []
for i in range(len(df.text)):
  text = df.text[i]
  z = sent_tokenize(text)
  lis.append(z)
```

```python
lis
```

```python
# for first 4 and last 4 senteces 
main = []
for i in range(len(lis)):
  t=[]
  l = len(lis[i])
  a = lis[i][0]
  b = lis[i][1]
  c = lis[i][2]
  d = lis[i][3]
  e = lis[i][l-4]
  f = lis[i][l-3]
  g = lis[i][l-2]
  h = lis[i][l-1]
  t.extend((a,b,c,d,e,f,g,h))
  main.append(t)
```

```python
main = lis
```

```python
hypothesis_template = 'This text is about {} .'
```

```python
from transformers import pipeline
zero_shot_classifier = pipeline("zero-shot-classification",model="facebook/bart-large-mnli")
```

```python
listt = []
for i in range(len(df.title)):
  sequence = ["{s}".format(s=df.title[i])]
  #candidate_labels = ["Crime", "Politics", "National", "Information Warfare","Threats","Cyber Security","Government","International"]
  candidate_labels = ["Crime", "Politics", "Terrorism", "Finance","Sports","Entertainment"]  
  poping_sequence = zero_shot_classifier(sequence, candidate_labels,hypothesis_template=hypothesis_template, multi_label=True)
  listt.append(poping_sequence)
  #print(poping_sequence)
```

```python
listt
```

```python
l_a=[]
l_b=[]
l_c=[]
l_d=[]
l_e=[]
l_f=[]
l_g=[]
l_h=[]
l_i=[]
l_j=[]
abc=[]
dedf=[]
for j in range(len(listt)): 
    a=listt[j][0]['labels'][0]
    b=listt[j][0]['scores'][0]
    c=listt[j][0]['labels'][1]
    d=listt[j][0]['scores'][1]
    e=listt[j][0]['labels'][2]
    f=listt[j][0]['scores'][2]
    g=listt[j][0]['labels'][3]
    h=listt[j][0]['scores'][3]
    ee=listt[j][0]['labels'][4]
    fe=listt[j][0]['scores'][4]
    ge=listt[j][0]['labels'][5]
    he=listt[j][0]['scores'][5]
    l_a.append(a)
    l_b.append(b)
    l_c.append(c)
    l_d.append(d)
    l_e.append(e)
    l_f.append(f)
    l_g.append(g)
    l_h.append(h)
    l_i.append(ee)
    l_j.append(fe)
    abc.append(ge)
    dedf.append(he)
```

```python
data_title = {'title':df.title,'highest label':l_a,'score1':l_b,'second label':l_c,'score2':l_d,'third label':l_e,'score3':l_f,'fourth label':l_g,'score4':l_h,'fifth label':l_i,'score5':l_j,'sixth label':abc,'score6':dedf}
df_title = pd.DataFrame(data_title)
```

```python
n = []
for i in range(len(four_list)):
  s = []
  for j in range(len(four_list[i])):
      dict = {}
      dict['{}'.format(four_list[i][j]['labels'][0])] = four_list[i][j]['scores'][0]
      dict['{}'.format(four_list[i][j]['labels'][1])] = four_list[i][j]['scores'][1]
      dict['{}'.format(four_list[i][j]['labels'][2])] = four_list[i][j]['scores'][2]
      dict['{}'.format(four_list[i][j]['labels'][3])] = four_list[i][j]['scores'][3]
      dict['{}'.format(four_list[i][j]['labels'][4])] = four_list[i][j]['scores'][4]
      dict['{}'.format(four_list[i][j]['labels'][5])] = four_list[i][j]['scores'][5]
      s.append(dict)
  n.append(s)
n
```

```python
len(n)
```

```python
ss = []
for i in range(len(n)):
  result = {}
  for elm in n[i]:
      for k, v in elm.items():
          if k not in result:
              result[k] = 0
          result[k] += v
  ss.append(result)
```
